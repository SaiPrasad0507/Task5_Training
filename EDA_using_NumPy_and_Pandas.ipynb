# What is Exploratory Data Analysis (EDA)?
EDA is the process of analyzing datasets to understand their structure, identify patterns, and extract useful insights.

EDA helps to:

~ Understand dataset structure

~ Find missing values

~ Analyze statistics

~ Visualize patterns in data

# Step 11: NumPy Statistical Analysis

```python
numeric_data = df.select_dtypes(include=np.number)

print("Mean Values")
print(np.mean(numeric_data))

print("\nMedian Values")
print(np.median(numeric_data))

print("\nStandard Deviation")
print(np.std(numeric_data))
```

Observation

Provides statistical insights using NumPy.

---



# Step 12: Correlation Analysis

```python
corr = numeric_data.corr()
print(corr)
```

Observation

Shows relationships between scores, wickets, and player performance.

---



# Step 13: Team-wise Wins Analysis

```python
winner_counts = df["Winner"].value_counts()

print(winner_counts)
plt.figure(figsize=(10,5))

winner_counts.plot(kind='bar')

plt.title("IPL Team Wins Distribution")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")

plt.show()
```

Observation

Identifies the most successful team.

---



# Step 14: Venue-wise Match Distribution

```python
venue_counts = df["Venue"].value_counts()

print(venue_counts)
plt.figure(figsize=(10,5))

venue_counts.plot(kind='bar')

plt.title("Matches Played at Different Venues")
plt.xlabel("Venue")
plt.ylabel("Matches")

plt.show()
```

Observation

Shows venues hosting the highest number of matches.

---



# Step 15: Team 1 Score Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Team1_Score"], bins=10)

plt.title("Team 1 Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```

Observation

Displays scoring trends of Team 1.

---



# Step 16: Team 2 Score Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Team2_Score"], bins=10)

plt.title("Team 2 Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```

Observation

Displays scoring trends of Team 2.

---



# Step 17: Team 1 vs Team 2 Scores

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Team1_Score"],
    df["Team2_Score"]
)

plt.title("Team1 Score vs Team2 Score")
plt.xlabel("Team1 Score")
plt.ylabel("Team2 Score")

plt.show()
```

Observation

Shows relationship between scores of competing teams.

---



# Step 18: Wickets Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df["Team1_Wickets"], bins=6)

plt.title("Team1 Wickets Distribution")
plt.xlabel("Wickets")
plt.ylabel("Frequency")

plt.show()

plt.figure(figsize=(8,5))

plt.hist(df["Team2_Wickets"], bins=6)

plt.title("Team2 Wickets Distribution")
plt.xlabel("Wickets")
plt.ylabel("Frequency")

plt.show()
```

Observation

Analyzes wicket-taking trends.

---



# Step 19: Player of Match Performance

```python
plt.figure(figsize=(8,5))

plt.hist(df["Player_of_Match_Runs"], bins=10)

plt.title("Player of Match Runs Distribution")
plt.xlabel("Runs")
plt.ylabel("Frequency")

plt.show()
```

Observation

Shows performance levels of Player of the Match winners.

---



# Step 20: Average Scores by Venue

```python
avg_scores = df.groupby("Venue")[["Team1_Score","Team2_Score"]].mean()

avg_scores

avg_scores.plot(kind='bar', figsize=(10,5))

plt.title("Average Scores by Venue")
plt.xlabel("Venue")
plt.ylabel("Average Score")

plt.show()
```

Observation

Identifies batting-friendly venues.

---



# Step 21: Top 5 Highest Scoring Matches

```python
df["Total_Score"] = df["Team1_Score"] + df["Team2_Score"]

top_matches = df.sort_values(
    by="Total_Score",
    ascending=False
)

top_matches[
    ["Match_ID","Team1","Team2","Total_Score"]
].head()
```

Observation

Displays highest-scoring matches.
